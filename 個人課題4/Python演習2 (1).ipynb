## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
#  図形の大きさを決める変数を定義
size = 5
# 2. for文を用いてループ処理を行う
# range(1, size + 1) とすることで、1からsizeまでの数字を順番に扱う
for i in range(1, size + 1):
    # アスタリスクをi回繰り返して表示
    print('*' * i)


````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
#  図形の大きさを決める変数を定義
size = 5
# for文を使って1行ずつ処理
for i in range(size):
    # i個のスペースの後に、1つのアスタリスクを表示
    print(' ' * i + '*')
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
# 1. 図形の大きさを決める変数を定義
size = 5
# for文を使って1行ずつ処理
for i in range(size):
    # (size - i - 1)個のスペースの後に、1つのアスタリスクを表示
    print(' ' * (size - i - 1) + '*')
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
# 図形の大きさを決める変数を定義
size = 5
# 二重のfor文を使って行と列を処理
for i in range(size):
    for j in range(size):
        # 左上から右下への対角線 (i == j)
        # 右上から左下への対角線 (i + j == size - 1)
        if i == j or i + j == size - 1:
            print('*', end='')
        else:
            print(' ', end='')
    # 1行描き終えたら改行
    print()
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
# 図形の大きさを決める変数を定義
size = 5
# for文を使って1行ずつ処理
for i in range(size):
    # 1. 左側の空白を計算（下の段に行くほど減る）
    spaces = ' ' * (size - i - 1)
    
    # 2. アスタリスクの数を計算（1, 3, 5, 7... と増える）
    # 数式: 2 * i + 1
    stars = '*' * (2 * i + 1)
    
    # まとめて表示
    print(spaces + stars)
````
